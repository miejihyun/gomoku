# Battle 1

UCB vs RAVE

- black: UCB
- white: RAVE

```python
win_black = 0
win_white = 0
winner = []
for i in range(10):
    board = Board()
    terminal = False
    while not terminal:
        if board.turn == White:
            move = BestMoveRAVE(board,10000)
            board.play(move)
        else:
            move = UCB(board,10000)
            board.play(move)
        terminal = board.terminal()
        if terminal:
            if board.score() == 1:
                win_white += 1
            if board.score() == 0:
                win_black += 1
            winner.append(board.score())
```

```python
print(winner)
```
